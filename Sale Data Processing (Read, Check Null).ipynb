```python
#membaca data customers, melakukan pengecekan data pada kolom customer_state
customers.head(5)
```

```python
#mengubah value pada kolom customer_state dari abreviation menjadi full name state

area_replace = {"AC":"Acre","AL":"Alagoas","AP":"Amapá","AM":"Amazonas","BA":"Bahia","CE":"Ceará","DF":"Distrito Federal","ES":"Espírito Santo","GO":"Goiás","MA":"Maranhão","MT":"MatoGrosso","MS":"MatoGrosso do Sul","MG":"Minas Gerais","PA":"Pará","PB":"Paraíba","PR":"Paraná","PE":"Pernambuco","PI":"Piauí","RJ":"Rio de Janeiro","RN":"Rio Grande do Norte","RS":"Rio Grande do Sul","RO":"Rondônia","RR":"Roraima","SC":"Santa Catarina","SP":"São Paulo","SE":"Sergipe","TO":"Tocantins",}
customers = customers.replace({'customer_state': area_replace})
customers.head(5)
```

```python
#mengecek bahwa semua data di kolom customer_state sudah berubah

customers['customer_state'].unique()
```

```python
#membaca data customers, melakukan pengecekan data pada kolom seller_state
sellers.head(5)
```
