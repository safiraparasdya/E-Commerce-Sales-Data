```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#membaca data
customers2 = pd.read_csv('clean_data/customers.csv')
order_items2 = pd.read_csv('clean_data/olist_order_items_dataset.csv')
reviews2 = pd.read_csv('clean_data/reviews.csv')
orders2 = pd.read_csv('clean_data/olist_orders_dataset.csv')
products2 = pd.read_csv('clean_data/translated_products.csv')
sellers2 = pd.read_csv('clean_data/sellers.csv')
```

```python
order_review = pd.merge(orders2,reviews2,how='inner',on='order_id')
```

```python
order_review
```

```python
order_review_cust = pd.merge(order_review,customers2,how='inner',on='customer_id')
```

```python
order_review_cust_orderitems = pd.merge(order_review_cust,order_items2,how='inner',on='order_id')
```

```python
order_review_cust_orderitems_sellers = pd.merge(order_review_cust_orderitems,sellers2,how='inner',on='seller_id')
```

```python
order_review_cust_orderitems_sellers_products = pd.merge(order_review_cust_orderitems_sellers,products2,
                                                         how='inner',on='product_id')
```
